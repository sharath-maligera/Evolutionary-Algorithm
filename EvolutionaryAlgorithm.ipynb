```python
import random
import math
import numpy as np
import scipy.stats

# function to generate initial population of 2 individuals where:
# each individual or genome is composed of 5 genes 
# each gene consists of real values from 0 to 5
def GeneratePopulation(noIndividuals, geneLength):   
    individual = ()    
    population = {}  
    for pop in range (0, noIndividuals):
        # to generate 5 genes having real values from 0 to 5
        individual = np.random.randint(5, size=geneLength)
        population[pop] = individual

    return (population) 


# function to calculate the fitness score of the individual or genome
# here fitness score is calculated as the sum of real values in genes
def calcFitness(individual):
    fitness = 0
    fitness = np.sum(individual)
    return fitness


# function to select the fittest of the individual from the genome population
# based on the fitness score
def getFittest(population,fitnessOfIndividual):
    # to calculate fitness of the fittest
    fittnessOfFittest = fitnessOfIndividual[max(fitnessOfIndividual, key=fitnessOfIndividual.get)]
    # to get the index of the fittest individual
    fittestInd = population[max(fitnessOfIndividual, key=fitnessOfIndividual.get)]
    return fittnessOfFittest,fittestInd


# function to inherit from the parent where:
# the contents of the surviving genome is copied into the newly created one
def inheritance(fittestIndividual):
    family = {}
    for f in range(0,2):
        family[f] = fittestIndividual
        
    return family

# function to perform mutation of the offspring
def mutation(offSpring):
    lower = -1
    upper = 1
    mu = 0.0
    sigma = 0.5
    N = 5
    # to generate 5 random vectors normally distributed between -1 to +1 and mean = 0, sigma = 0.5
    randomVector = scipy.stats.truncnorm.rvs((lower-mu)/sigma,(upper-mu)/sigma,loc=mu,scale=sigma,size=N)
    # adding random vector to all genes of the genome
    r = np.add(offSpring,randomVector)
    
    mutatedOffSpring = np.int_(np.ceil(abs(r)))
      
    return mutatedOffSpring
    
    
if __name__ == '__main__':
    
    # genome length L = 5
    genomeLength = 5
    
    # initial population of 2 individuals
    noIndividuals = 2
    
    # Maximum value of the fittnes function to be achieved
    MaximumFittness = 100
    
    # generate initial population
    population = GeneratePopulation(noIndividuals,genomeLength)
    
    iteration = 0
    
    # while maximum fitness is not reached
    while True:
        
        print("At iteration: {}".format(iteration))
        
        # calculate the fitness of each parents
        fitnessOfIndividuals = {}
        for i in range(len(population)):
            fitnessOfIndividuals[i] = calcFitness(population[i])
            print("fitness Of genome{} {} is : {}".format(i+1,str(population[i]),str(fitnessOfIndividuals[i])))
        
        # select the fittest parent for generation of offspring
        fittnessOfFittest,fittestIndividual = getFittest(population,fitnessOfIndividuals)
        
        # if the maximum fitness i.e 100 is reached; break
        if(fittnessOfFittest >= MaximumFittness):
            print("Max fittness is reached")
            print("contents of final genome is : {}".format(str(fittestIndividual)))
            break
        
        iteration += 1
        
        # inheritance phase where the surviving genome is copied to be the newly created genome 
        popOfFamily = inheritance(fittestIndividual)

        # newly created genome on which mutation has to be performed
        offSpring = popOfFamily[1]

        # mutation phase
        mutatedGenome = mutation(offSpring)
        
        # new population consisting of parent and mutated genome
        popOfFamily[1] = mutatedGenome
        population = popOfFamily
       
    
    
    
```
